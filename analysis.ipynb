```python
import pandas as pd
import os
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///inventory.db')
```

```python
def ingest_db(chunk, table_name, engine):
    chunk.to_sql(table_name, con=engine, if_exists='append', index=False)
```

```python
for file in os.listdir('data'):
    if file.lower().endswith('.csv'):
        print(f"\nProcessing: {file}")
        table_name = file[:-4]
        chunk_count = 0

        # Read in chunks and process each
        for chunk in pd.read_csv(os.path.join('data', file), chunksize=100_000):
            print(f"  - Ingesting chunk {chunk_count+1}, shape: {chunk.shape}")
            ingest_db(chunk, table_name, engine)
            chunk_count += 1

        print(f"Done with {file}: {chunk_count} chunks ingested.\n")
```
